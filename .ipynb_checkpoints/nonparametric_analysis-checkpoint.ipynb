## Part II. Number of Words per Sentence and Histograms

We first compute the number of words in each sentence for each of Cicero's works. We then pick a representative work from each of Cicero's three genres and look at the histograms of number of words per sentence for those works. His three genres are _philosophy_, _oratory_, and _epistolatory_.

```python
hist_data = {text: [len(sent) for sent in tokenized_texts[text]] for text in tokenized_texts}
```

```python
import numpy as np

arr1 = np.array(hist_data['repub']) # philosophy
arr2 = np.array(hist_data['cael'])  # oratory
arr3 = np.array(hist_data['att'])   # epistolatory
```

```python
%%R -i arr1 -i arr2 -i arr3 -w 10 -h 5 --units in -r 200
require(ggplot)

df = data.frame(legend="De Re Publica (Philosophy)", val=arr1)
df = rbind(df,data.frame(legend="Pro Caelio (Oratory)", val=arr2))
df = rbind(df, data.frame(legend="Ad Atticum (Epistolatory)", val=arr3))

ggplot(data=df, aes(x=val, fill=legend, color=legend)) + 
    geom_histogram(binwidth=2,alpha=0.5,aes(y=..density..),position="identity") + 
    facet_wrap(~legend) +
    theme(legend.position="none") + 
    xlab("Number of Words per Sentence")
```

It seems that the _De Re Publica_ and _Pro Caelio_ have more similar distributions to each other than with _Ad Atticum_. I suspect this is because Cicero, who is known for very eloquent and well-thought out texts, has put more time into developing philosophy and oratory works. The epistolary genre tends to contain more informal talk to friends and relatives, and is thus more likely to have shorter and simpler sentences. However, I wonder: are the texts I picked representative of each genre, or are they outliers?

This leads me to consider distributions for the number of words per sentence across all texts in each of Cicero's three genres.

```python
phil_texts = set(["inventione", "optgen", "topica", "oratore", "fato", "paradoxa", "partitione", "brut", "orator", "repub", "consulatu", "leg", "fin", "tusc", "nd", "acad", "senectute", "amic", "divinatione", "off", "compet"])
orat_texts = set(["quinc", "sex", "rosccom", "legagr", "ver", "imp", "caecina", "cluentio", "rabirio", "cat", "murena", "sulla", "flacco", "arch", "postreditum", "domo", "haruspicum", "plancio", "sestio", "vatin", "cael", "prov", "balbo", "milo", "piso", "scauro", "fonteio", "rabiriopost", "marc", "lig", "deio", "phil"])
epis_texts = set(["att", "fam", "fratrem", "adbrutum"])

arr4 = []
arr5 = []
arr6 = []
for text in hist_data:
    if text in phil_texts:
        arr4.extend(hist_data[text])
    elif text in orat_texts:
        arr5.extend(hist_data[text])
    elif text in epis_texts:
        arr6.extend(hist_data[text])
        
arr4 = np.array(arr4)
arr5 = np.array(arr5)
arr6 = np.array(arr6)
```

```python
%%R -i arr4 -i arr5 -i arr6 -w 10 -h 5 --units in -r 200
require(ggplot)

df2 = data.frame(legend="Philosophy", val=arr4)
df2 = rbind(df2,data.frame(legend="Oratory", val=arr5))
df2 = rbind(df2, data.frame(legend="Epistolatory", val=arr6))

ggplot(data=df2, aes(x=val, fill=legend, color=legend)) + 
    geom_histogram(binwidth=4,alpha=0.5,aes(y=..density..),position="identity") + 
    facet_wrap(~legend) +
    theme(legend.position="none") + 
    xlab("Number of Words per Sentence")
```

These distributions confirm the idea that Cicero's philosophy and oratory texts are more similar to each other than to his epistolary works.

## Part IV. Comparison of Distributions
Now I wish to compare the distributions of philosophy, oratory, and epistolatory. To compare histograms, I must choose an error metric. I investigate 4 error metrics:

**Correlation:** $d(H_1, H_2) = \frac{\sum_{bin} (H_1(bin) - \bar H_1)(H_2(bin) - \bar H_2)}{\sqrt{\sum_{bin} H_1(bin)^2} \cdot \sqrt{\sum_{bin} H_2(bin)^2}}$    where $\bar H_i = \frac{1}{N} \sum_{bin} H_i(bin)$    and $N$ is the number of bins

**Intersection:** $d(H_1, H_2) = \sum_{bin} \min\{(H_1(bin), H_2(bin)\}$

**Chi-Squared:** $d(H_1, H_2) = \sum_{bin} \frac{(H_1(bin) - H_2(bin))^2}{H_1(bin)}$

**Hellinger (Battacharyya):** $d(H_1, H_2) = \sqrt{1 - \frac{1}{\sqrt{\bar H_1 \bar H_2 N^2}} \sum_{bin} \sqrt{H_1(bin)\cdot H_2(bin)}}$

Note that for _correlation_ and _intersection_, a higher value indicates greater similarity, while for _chi-squared_ and _Hellinger_, a lower value indicates greater similarity.

```python
import cv2
import pandas as pd

histograms = [arr4, arr5, arr6]
max_val = max([max(x) for x in histograms])
histograms = list(map(lambda x: np.histogram(x, range=(0, max_val), bins = int(max_val/2), density=True), histograms)) # Fix bin width, make sure bins correspond
histograms = list(map(lambda x: x[0].ravel().astype('float32'), histograms))

cv_metrics = {"Correlation": cv2.HISTCMP_CORREL,
    "Intersection": cv2.HISTCMP_INTERSECT, 
    "Chi-Squared": cv2.HISTCMP_CHISQR,
    "Hellinger": cv2.HISTCMP_BHATTACHARYYA}

results = {}
for metric in cv_metrics:
    mat = np.zeros((3,3))
    
    for i in range(len(histograms)):
        for j in range(len(histograms)):
            min_max_bin = min(len(histograms[i]), len(histograms[j])) # Just compare up to the highest bin in the smaller histogram
            mat[i,j] = cv2.compareHist(histograms[i][0:min_max_bin], histograms[j][0:min_max_bin], cv_metrics[metric])
    
    results[metric] = pd.DataFrame(mat, columns=["Philosophy", "Oratory", "Epistolatory"], index=["Philosophy", "Oratory", "Epistolatory"])

for metric in results:
    print(metric + ":")
    print(results[metric])
    print("\n")
```

As expected, all four comparison metrics lead us to believe that philosophy and oratory have much more similar distributions than with epistolatory. Interestingly, all four metrics also suggest that oratory is more similar to epistolatory, which is supported by a cursory glance at the frequency polygons above.

In conclusion, a nonparametric analysis of Cicero's texts can reveal much about the similarities and differences between his genres of writing. Histograms and frequency polygons help us visualize the distributions of the number of words per sentence, and a rigorous histogram comparison can numerically validate our claims about the similarities and differences between Cicero's styles of writing.
